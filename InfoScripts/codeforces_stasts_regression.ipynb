```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as md
import numpy as np
from datetime import datetime
import seaborn as sns
%matplotlib inline

stats = requests.get('http://codeforces.com/api/user.rating?handle=leontibrechko').json()['result']
data = pd.DataFrame.from_dict(stats)
```

```python
date = [date[0] for date in data.as_matrix(columns=['ratingUpdateTimeSeconds'])]
ratings = [rating[0] for rating in data.as_matrix(columns=['newRating'])]

f = np.poly1d(np.polyfit(date, ratings, 5))
fixed = lambda x: 1400;

x_pred = np.linspace(1.445e9, 1.480e9, 0.001e9)
ax = data.plot(x='ratingUpdateTimeSeconds', y='newRating', kind='scatter')
ax.set_ylim(1100, 1600)
plt.plot(x_pred, f(x_pred))
```
